```python
import re
import nltk
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from textblob import Word
```

```python
import pandas as pd
import subprocess
def curl_request(url):

    # Define the command to execute using curl
    command = ['curl', '-s', '-o', '-', url]

    # Execute the curl command and capture the output
    result = subprocess.run(command, capture_output=True, text=True)

    # Return the stdout of the curl command
    return result.stdout

# Make a curl request to https://www.google.com/
response = curl_request('https://datasets-server.huggingface.co/rows?dataset=dair-ai%2Femotion&config=split&split=train&offset=0&limit=100')

# df = pd.DataFrame(response)
```

```python
import json

data = json.loads(response)
rows = data['rows']
labels = [row['row']['label'] for row in rows]
texts = [row['row']['text'] for row in rows]
df = pd.DataFrame({'text': texts, 'label': labels})
```

```python
df.head()
```

```python
X = df['text']  # Assuming 'text_column' contains the text data
y = df['label']  # Assuming 'emotion_column' contains the emotion labels

# Map numerical labels to emotions
label_mapping = {0: 'sadness', 1: 'joy', 2: 'love', 3: 'anger', 4: 'fear', 5: 'surprise'}
y = y.map(label_mapping)
```

```python
df['emotion'] = y
```

```python
df = df.drop(columns = ['label'])
```
